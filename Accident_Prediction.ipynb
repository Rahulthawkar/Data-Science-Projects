OneHot Encoding

```python
col = ['road_type','lighting','weather',
       'road_signs_present','public_road',
       'time_of_day','holiday','school_season']

# Use one-hot encoding for categorical variables if not already done
if set(col).issubset(df.columns):
    df = pd.get_dummies(df, columns=col, drop_first=True)

# Check the transformed dataframe
df.head()
```

```python
x= df.drop('accident_risk',axis=1).values
y= df['accident_risk'].values
```

```python
y.shape
```

```python
x.shape
```

### Now we split our dataset as train set and test set.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                   test_size=0.20, random_state=0)
```

Feature Scalingj

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
import matplotlib.pyplot as plt
plt.plot(y_test, color='blue', label='Actual')
plt.plot(y_pred, color='red', label='Predicted')
plt.legend()
plt.show()
```

Evaluation Techniques

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)  
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```
